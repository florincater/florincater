```python

DC = M.iloc[:,3:14]
p1 = DC.shape[1]
print(DC)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
X= scaler.fit_transform(DC)
print(X.round(3))
None
```

```python
corr_matrix = np.corrcoef(X.transpose())
print("La matrice de corrélation: \n", corr_matrix.round(2))
```

```python
#En se basant sur le critère de Kaiser, combien de composantes principales devrons-nous retenir ?
# Calcul de valeurs propres et des vecteurs propres

eigval, vectors = np.linalg.eig(corr_matrix )
print('Les valeurs propres sont:', '\n', eigval.round(3), '\n')
# Selon Kaiser on peut retenir 2 composantes principales
```

```python
from sklearn.decomposition import PCA
# Instanciation
# Le paramètre (svd_solver = ‘full’) indique l’algorithme
# utilisé pour la décomposition en valeurs singulières.
# La méthode ‘’full’’ est sélectionnée par défaut pour
# le traitement de bases de données de taille réduite.
# Le nombre de composantes n_components n’étant pas spécifié, il est fixé
# par défaut au nombre de variables (n_components=K par défaults).
acp = PCA(svd_solver='full', n_components=None)

C=acp.fit_transform(X)
print(C.round(3))
```

```python
# Représentez les individus sur les deux premières composantes principales.
fig, axes = plt.subplots(figsize=(12,8))
axes.set_xlim(-6,6) #même limites en abscisse
axes.set_ylim(-2,3) #et en ordonnée
#placement des étiquettes des observations
for i in range(N):
 plt.annotate(M.index[i],(C[i,0],C[i,1]))
#ajouter les axes
plt.plot([-6,6],[0,0],color='silver',linestyle='-',linewidth=1)
plt.plot([0,0],[-2,3],color='silver',linestyle='-',linewidth=1)
plt.ylabel("Axe factoriel 2 (14%)")
plt.xlabel("Axe factoriel 1 (74%)" )
#affichage
plt.show()
```
